```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        result = -1

        while low <= high:
            mid = (low + high) // 2
            if arr[mid] == k:
                return mid
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1
        return result
```

```python
if __name__ == "__main__":
    import numpy as np
    arr = sorted(np.random.randint(7, size=5))
    k = np.random.randint(0, 6)
    solution = Solution()
    index_pos = solution.binarysearch(arr, k)

    if index_pos != -1:
        print(f"The index position of {k} is: {index_pos}")
    else:
        print(f"{k} is not present in the array.")
```
