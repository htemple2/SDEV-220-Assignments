```python
from sqlalchemy import create_engine

engine = create_engine("sqlite:///books.db")

try:
    connection = engine.connect()
    print("Successfully connected to books.db using SQLAlchemy.")
    connection.close()
except Exception as e:
    print(f"Connection failed: {e}")
```

```python
import sqlite3
import csv
import os

db_file = 'books.db'
csv_file = 'books2.csv'

if not os.path.exists(csv_file):
    print(f"Error: {csv_file} not found. Please create it first.")
else:
    try:
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()

        with open(csv_file, 'r', newline='', encoding='utf-8') as file:

            books = csv.DictReader(file)

            next(books)

            insert_query = '''
                INSERT INTO books (title, author, year)
                VALUES (:title, :author, :year)
            '''

            cursor.executemany(insert_query, books)

            conn.commit()
            print(f"Successfully inserted data from {csv_file} into books table.")

    except sqlite3.Error as e:
        print(f"An SQLite error occurred: {e}")
    except IOError as e:
        print(f"An IO error occurred: {e}")
    finally:
        if conn:
            conn.close()
```
